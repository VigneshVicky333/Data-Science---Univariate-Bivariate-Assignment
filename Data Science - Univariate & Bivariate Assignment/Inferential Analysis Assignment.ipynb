# Analysis of Variance Between etest_p and mba_p at 5% Significance Level

### Method:
**One-Way ANOVA** is appropriate for this analysis.

---

### Explanation:

#### Why One-Way ANOVA?
- You are testing whether the means of **mba_p** differ significantly across the levels of **etest_p**.
- This involves:
  - One dependent variable (**mba_p**).
  - One independent categorical variable (**etest_p**) or binning **etest_p** if it is continuous.

#### Why Not Two-Way Classification?
- Two-Way ANOVA is used when analyzing two independent categorical variables and their interaction on a dependent variable.
- Here, you have only one independent variable (**etest_p**), so One-Way ANOVA is more suitable.


````python
# Here we are going to check for entrance-Pass mark and mba-Pass mark(going to check for 2 columns only) so we can use one way Classification:

import scipy.stats as stats

stats.f_oneway(dataset['etest_p'],dataset['mba_p'])
````

---

### Decision Rule:
- **If p < 0.05:** Reject Null Hypothesis (significant difference exists).
- **If p > 0.05:** Accept Null Hypothesis (no significant difference exists).

---

### Conclusion:
- **P-value = 2.7511e-16**, which is much smaller than **0.05**.
- **Decision:** Reject the Null Hypothesis (H₀).
- **Interpretation:** There is a significant difference in means between etest_p and mba_p.

# 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.
# (Make decision using Hypothesis Testing)

# Hypothesis Testing: Relationship Between Degree Technology, Specialisation, and Salary

## Appropriate Method: **Two-Way ANOVA (Two-Way Classification)**

- To check the relationship between **degree technology (Sci&Tech)**, **specialisation (Mkt&HR)**, and **salary** at a significance level of 5%, Two-Way ANOVA is the appropriate hypothesis testing method.

---

## Why Not One-Way Classification?

- **One-Way ANOVA** tests only one categorical independent variable at a time.  
- Since we are analyzing the effects of **two categorical variables** (**degree_t** and **specialisation**) and their **interaction** on salary, **Two-Way ANOVA** is more suitable.

---

### Conclusion:
Two-Way ANOVA allows us to test:
1. The **main effects** of each independent variable (degree_t and specialisation).
2. The **interaction effect** between the independent variables on salary.

This method provides a comprehensive understanding of how these variables influence salary.


````python
# To check the relationship between degree technology (sci&Tech), specialisation (Mkt&HR), and salary at a significance level of 5%,
# Two-Way ANOVA (Two-Way Classification) is the appropriate hypothesis testing method.

# Why Not One-Way Classification?
# One-Way ANOVA tests only one categorical independent variable at a time. 
# Since we are analyzing the effects of two categorical variables (degree_t and specialisation) and their interaction on salary, 
# Two-Way ANOVA is more appropriate.

import statsmodels.api as sm
from statsmodels.formula.api import ols

# Perform Two-Way ANOVA:

model = ols('salary ~ C(degree_t) + C(specialisation) + C(degree_t):C(specialisation)', data=dataset).fit()
anova_results = sm.stats.anova_lm(model, typ=2)
````

````python
# Display the results
print(anova_results)
````

# 9)Convert the normal distribution to standard normal distribution for salary column

````python
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
````

### Explanation of the Code:

#### 1. **Importing Library**:
- `import seaborn as sns`: Import the Seaborn library for visualization.

#### 2. **Calculate Mean and Standard Deviation**:
- `mean = dataset.mean()`: Calculates the average value of the dataset.
- `std = dataset.std()`: Calculates the spread of the dataset.

#### 3. **Standardize the Dataset**:
- `values = [i for i in dataset]`: Creates a list of all values in the dataset.
- `z_score = [((j - mean) / std) for j in values]`: 
  - This is a **one-liner for loop** that calculates the Z-score for each value in the dataset.
  - **Z-score formula**: 
    \[
    Z = \frac{X - \mu}{\sigma}
    \]
    Where:
    - \( X \): Value in the dataset.
    - \( \mu \): Mean of the dataset.
    - \( \sigma \): Standard deviation of the dataset.

#### 4. **Plot the Standardized Data**:
- `sns.distplot(z_score, kde=True)`:
  - Generates a distribution plot for the Z-scores using Seaborn.
  - The parameter `kde=True` ensures the graph includes a Kernel Density Estimate (smooth curve) along with the histogram.

#### 5. **Return Mean of Z-Scores**:
- `sum(z_score) / len(z_score)`:
  - Calculates the mean of the Z-scores.
  - For a standard normal distribution, this value should be close to 0.

### Summary:
This code:
1. Calculates the Z-scores for a dataset to standardize it.
2. Visualizes the standardized data using Seaborn's `distplot`.
3. Verifies the standardization by calculating the mean of the Z-scores, which should ideally be 0.

# Explanation of `stdNBgraph` function for converting data to a Standard Normal Distribution:

This function converts a given dataset to a standard normal distribution by calculating the Z-scores for each value. The Z-score represents how many standard deviations away a data point is from the mean.

### Code Breakdown:

```python
def stdNBgraph(dataset):
    # Convert to Standard Normal Distribution
    import seaborn as sns
    mean = dataset.mean()  # Calculate the mean of the dataset
    std = dataset.std()    # Calculate the standard deviation of the dataset
    
    # Create a list of the dataset values
    values = [i for i in dataset]
    
    # Calculate the Z-score for each value in the dataset
    z_score = [((j - mean) / std) for j in values]
    
    # Visualize the distribution of Z-scores (standardized data)
    sns.distplot(z_score, kde=True)
    
    # The following line computes the mean of the Z-scores, which should ideally be close to 0 for a normal distribution
    sum(z_score) / len(z_score)
    
    # Alternatively, you can calculate the standard deviation of Z-scores to verify if it is close to 1
    # z_score.std()


### Key Steps Explained:

#### 1. **Calculating Mean and Standard Deviation**:
- `mean = dataset.mean()`: This calculates the average value of the dataset. It serves as the center of the distribution.
- `std = dataset.std()`: This calculates the spread of the dataset, indicating how much the values deviate from the mean.

#### 2. **Z-Score Calculation**:
- `z_score = [((j - mean) / std) for j in values]`: This step standardizes the data by calculating the Z-score for each value.
- The formula for the Z-score is:
  
  \[
  Z = \frac{X - \mu}{\sigma}
  \]

  Where:
  - \( X \) is the value in the dataset.
  - \( \mu \) is the mean of the dataset.
  - \( \sigma \) is the standard deviation of the dataset.

````python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
````

# Question No: 11) Test the similarity between the degree_t (Sci&Tech) with respect to etest_p and mba_p at a significance level of 5%. (Make decision using Hypothesis Testing)


For your specific case, where you are testing the similarity between degree_t (Sci&Tech) with respect to etest_p and mba_p at a significance level of 5%, the appropriate ANOVA test to use is **One-Way ANOVA**.

### Here's why One-Way ANOVA is more suitable:
One-Way ANOVA is used when you are comparing the means of more than two groups (categories) based on a single independent variable (factor). In your case, the factor you're concerned with is degree_t (Sci&Tech), and you're testing it against two different dependent variables (etest_p and mba_p).

You are interested in whether degree_t (i.e., Sci&Tech or other degree types) has a significant impact on the two test scores (etest_p and mba_p).

### Hypothesis Testing Setup:
- **Null Hypothesis (H₀):** There is no significant difference in test scores (etest_p or mba_p) based on the degree type (Sci&Tech).
- **Alternative Hypothesis (H₁):** There is a significant difference in test scores (etest_p or mba_p) based on the degree type (Sci&Tech).

### One-Way ANOVA Procedure:
- **Factor:** degree_t (Sci&Tech vs. other degrees)
- **Dependent Variables:** etest_p and mba_p (the test scores you are comparing based on degree type)


````python
# Code for One-Way ANOVA (for etest_p and degree_t):
````

````python
# Step 1: Load your actual dataset
````

````python
# Step 2: Check the structure of your dataset (optional, just to confirm)
print(dataset.head())  # This will display the first few rows of your dataset
````

````python
# Step 3: Proceed with the One-Way ANOVA code (no changes needed for this part)
from scipy import stats
````

````python
# One-Way ANOVA for etest_p with respect to degree_t
f_stat_1, p_value_1 = stats.f_oneway(dataset[dataset['degree_t'] == 'Sci&Tech']['etest_p'],
                                     dataset[dataset['degree_t'] != 'Sci&Tech']['etest_p'])
````

````python
# One-Way ANOVA for mba_p with respect to degree_t
f_stat_2, p_value_2 = stats.f_oneway(dataset[dataset['degree_t'] == 'Sci&Tech']['mba_p'],
                                     dataset[dataset['degree_t'] != 'Sci&Tech']['mba_p'])
````

````python
# Print the results
print(f"ANOVA for etest_p: F-statistic = {f_stat_1}, p-value = {p_value_1}")
print(f"ANOVA for mba_p: F-statistic = {f_stat_2}, p-value = {p_value_2}")
````

````python
# Make decisions
alpha = 0.05  # significance level
if p_value_1 < alpha:
    print("Reject null hypothesis for etest_p: Significant difference in test scores based on degree_t.")
else:
    print("Fail to reject null hypothesis for etest_p: No significant difference in test scores based on degree_t.")

if p_value_2 < alpha:
    print("Reject null hypothesis for mba_p: Significant difference in test scores based on degree_t.")
else:
    print("Fail to reject null hypothesis for mba_p: No significant difference in test scores based on degree_t.")
````

   - **`fmt='.2f'`**: This specifies the format of the annotations. `'.2f'` means that the correlation values will be displayed with two decimal places.
